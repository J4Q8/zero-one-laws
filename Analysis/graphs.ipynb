```python
import pandas as pd
import numpy as np
import plotly
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import kaleido
```

```python
df = pd.read_excel(r"C:\Users\jakub\Documents\Bachelor AI\Bachelor Thesis\Code\Analysis\dataset.xlsx")
```

```python
language = "GL"
mode = "model"

# dist_data=[iris_df[iris_df['class']=='Iris-setosa']['sepal_length'], iris_df[iris_df['class']=='Iris-versicolor']['sepal_length'],
#           iris_df[iris_df['class']=='Iris-virginica']['sepal_length']]

grop_labels=list(range(40, 81, 8))

dist_data = [df[f"{mode}_{language}_{node}"].fillna(0) for node in grop_labels]
```

```python
fig=go.Figure()

fig = ff.create_distplot(dist_data,grop_labels,show_hist=False)

fig.update_traces(opacity=0.55)
fig.update_layout(legend=dict(title="Number of worlds", yanchor="top",y=0.98,xanchor="right",x=0.95),
                  autosize=True, xaxis_title='Valid in x models out of 5000', yaxis_title='Relative frequency of formulas valid in x models out of 5000',
                  font=dict(size=20, family='Times New Romans', color='black'),
                  width=900, height=800
                 )

fig.update_xaxes(showline=True, linewidth=1, linecolor='white', gridwidth=3, gridcolor='white', mirror=True)
fig.update_yaxes(showline=True, linewidth=1, linecolor='white', gridwidth=3, gridcolor='white', mirror=True)

fig.show()
plotly.io.write_image(fig, f"{language}{mode}graph.png")
```
