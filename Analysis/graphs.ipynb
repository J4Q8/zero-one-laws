```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
df = pd.read_excel(r"C:\Users\jakub\Documents\Bachelor AI\Bachelor Thesis\Code\Analysis\dataset.xlsx")
```

```python
fig=go.Figure()

language = "GL"
mode = "frame"

# dist_data=[iris_df[iris_df['class']=='Iris-setosa']['sepal_length'], iris_df[iris_df['class']=='Iris-versicolor']['sepal_length'],
#           iris_df[iris_df['class']=='Iris-virginica']['sepal_length']]

grop_labels=list(range(40, 81, 8))

dist_data = [df[f"{mode}_{language}_{node}_exact"].fillna(0) for node in grop_labels]

fig = ff.create_distplot(dist_data,grop_labels,show_hist=False)

fig.update_traces(opacity=0.55)
fig.update_layout(title={'text': '0-1','y':0.95,'x':0.5, 'xanchor': 'center','yanchor': 'top'},
                          legend=dict(title="Structure sizes", yanchor="top",y=0.98,xanchor="right",x=0.95),
                  autosize=True,margin=dict(t=70,b=0,l=0,r=0), xaxis_title='Valid in n structures', yaxis_title='Frequency',
                  font=dict(size=20, family='Times New Romans', color='brown'),
                  width=900, height=800
                 )

fig.update_xaxes(showline=True, linewidth=1, linecolor='white', gridwidth=3, gridcolor='white', mirror=True)
fig.update_yaxes(showline=True, linewidth=1, linecolor='white', gridwidth=3, gridcolor='white', mirror=True)

fig.show()
```
